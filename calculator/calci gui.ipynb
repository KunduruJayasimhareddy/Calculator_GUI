```python
from tkinter import *

root=Tk()

root.title("Simple Calculator")
photo = PhotoImage(file = "calci.png")
root.iconphoto(False, photo)
root.geometry('210x240')

e=Entry(root,bd=3)

e.grid(column=0,row=0,columnspan=3,padx=7,pady=7)

def onClick(num):
   
    current=e.get()
    e.delete(0,END)
    e.insert(0,current+num)
    
def clear():
    e.delete(0,END)
    
def add():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="add"
    e.delete(0,END)

def sub():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="sub"
    e.delete(0,END)

def mul():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="mul"
    e.delete(0,END)

def div():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="div"
    e.delete(0,END)
    
def power():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="pow"
    e.delete(0,END)
    
def percent():
    first_number=e.get()
    global f_num
    f_num=float(first_number)
    global math
    math="percent"
    e.delete(0,END)
    
def equal():
    second_number=e.get()
    e.delete(0,END)
    
    if(math=="add"):
        e.insert(0,f_num+float(second_number))
    if(math=="sub"):
        e.insert(0,f_num-float(second_number))
    if(math=="mul"):
        e.insert(0,f_num*float(second_number))
    if(math=="div"):
        if(second_number == "0"):
            error="ERROR!"
            e.insert(0,error)
        else:
            e.insert(0,f_num/float(second_number))
    if(math=="pow"):
        e.insert(0,f_num**float(second_number))
    if(math=="percent"):
        e.insert(0,(f_num*float(second_number))/100)

button_1=Button(root,text="1",bg='#cdf8fa',command=lambda:onClick("1"))
button_2=Button(root,text="2",bg='#cdf8fa',command=lambda:onClick("2"))
button_3=Button(root,text="3",bg='#cdf8fa',command=lambda:onClick("3"))
button_4=Button(root,text="4",bg='#cdf8fa',command=lambda:onClick("4"))
button_5=Button(root,text="5",bg='#cdf8fa',command=lambda:onClick("5"))
button_6=Button(root,text="6",bg='#cdf8fa',command=lambda:onClick("6"))
button_7=Button(root,text="7",bg='#cdf8fa',command=lambda:onClick("7"))
button_8=Button(root,text="8",bg='#cdf8fa',command=lambda:onClick("8"))
button_9=Button(root,text="9",bg='#cdf8fa',command=lambda:onClick("9"))
button_0=Button(root,text="0",bg='#cdf8fa',command=lambda:onClick("0"))
button_dot=Button(root,text=".",padx=3,bg='#cdf8fa',command=lambda:onClick("."))


button_clear=Button(root,text="clear",bg='#cdf8fa',padx=25,command=clear)
button_equal=Button(root,text="=",bg='#cdf8fa',padx=35,command=equal)

button_add=Button(root,text="+",bg='#cdf8fa',command=add)
button_sub=Button(root,text="-",bg='#cdf8fa',command=sub)
button_mul=Button(root,text="x",bg='#cdf8fa',command=mul)
button_div=Button(root,text="/",bg='#cdf8fa',command=div)
button_pow=Button(root,text="^",bg='#cdf8fa',command=power)
botton_percent=Button(root,text="%",bg='#cdf8fa',command=percent)

button_7.grid(column=0,row=1)
button_8.grid(column=1,row=1)
button_9.grid(column=2,row=1)

button_4.grid(column=0,row=2)
button_5.grid(column=1,row=2)
button_6.grid(column=2,row=2)

button_1.grid(column=0,row=3)
button_2.grid(column=1,row=3)
button_3.grid(column=2,row=3)

button_add.grid(column=0,row=4)
button_0.grid(column=1,row=4)
button_dot.grid(column=2,row=4)

button_sub.grid(column=0,row=5)
button_equal.grid(column=1,row=5,columnspan=2)

button_mul.grid(column=0,row=6)
button_clear.grid(column=1,row=6,columnspan=2)

button_div.grid(column=0,row=7)
button_pow.grid(column=1,row=7)
botton_percent.grid(column=2,row=7)


root.mainloop()
```
